# Text preprocessing 

```python
import string
import nltk

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from nltk.stem import PorterStemmer

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
punctuation = string.punctuation
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
stemmer = PorterStemmer()
```

```python
def clean_text(text):
    text = text.lower()
    text = text.translate(str.maketrans('', '', string.punctuation))
    
    words = nltk.word_tokenize(text)
    
    words = [stemmer.stem(word) for word in words
             if word not in stop_words]
    
    return " ".join(words)
```

```python
data["clean_questions"] = data["question"].apply(clean_text) # apply clean_question to data)
```

# Feature Extraction (TfidfVectorizer / TF-IDF)

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
```

```python
vectorizer = TfidfVectorizer()

x = vectorizer.fit_transform(data["clean_questions"])
y = data["intent"]
```

```python
model = LogisticRegression()
model.fit(x, y)
```

```python
import random

def get_response(intent):
    responses = data[data["intent"] == intent]["response"].values
    return random.choice(responses)
```

# chatbot

```python
print("🤝welcom to our chatbot🛒 for your doubt and clarifications❤️")

print("🎉Chatbot ready! Type 'exit' to stop❤️")
while True:
    user_input = input("You: ")

    if user_input.lower() == "exit":
        print("🛒Bot: 👋❤️Goodbye!")
        break

    # clean user text
    cleaned = clean_text(user_input)

    # convert to vector
    user_vector = vectorizer.transform([cleaned])

    # predict intent
    predicted_intent = model.predict(user_vector)[0]

    # get response
    reply = get_response(predicted_intent)

    print("🛒Bot:", reply)
print("🎉thank you for your effort😊🌙❤️")
```
